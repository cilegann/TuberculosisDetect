```python
from PIL import Image
import numpy as np
from random import randint, uniform,choice
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
mapping_file='./mapping/190410_newdata_train_cnn_mapping.csv'
with open(mapping_file) as file:
    lines=file.readlines()
```

```python
x=[]
y=[]
for line in lines:
    line=line.replace("\n","")
    file,label=line.split(",")
    if label=='1':
        x.append(np.asarray(Image.open(file)))
        y.append(int(label))
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

```python
xx=np.reshape(x,(len(x),-1))
```

```python
N=10
k=5
```

```python
xx=np.reshape(x,(len(x),-1))
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(n_neighbors=k,p=2)
knn.fit(xx)
```

```python
for i in [259]:#range(len(x)):
    idx=knn.kneighbors([xx[i]], return_distance=False)
    print(idx)
    idx=idx[0][1:]
    for n in range(N):
        id=choice(idx)
        r=uniform(0,1)
        newImg=((1-r)*x[i]+r*(x[id])).astype(int)
        print("========"+str(r)+"=======")
        plt.imshow(x[i])
        plt.show()
        plt.imshow(x[id])
        plt.show()
        plt.imshow(newImg)
        plt.show()
```
