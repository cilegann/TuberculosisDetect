```python
database='./data'
```

```python
def getGroup(str):
    return str[:str.find("_")]
```

```python
def getDateStr(str):
    if '/' in str:
        str=str[str.rfind('/'):]
    return str[str.find("_")+1:str.find(".")]
```

```python
def getDate(str):
    return datetime.strptime(getDateStr(str), "%Y%m%d").date()
```

# Count Data pieces by Day

```python
dateData={}
for r,ds,fs in os.walk(database):
    for f in fs:
        path=os.path.join(r,f)
        d=getDateStr(path)
        if d not in dateData:
            dateData[d]=[0,0,0,0]
        if 'negative' in path:
            dateData[d][0]+=1
            dateData[d][3]+=1
        if 'positive' in path:
            dateData[d][1]+=1
            dateData[d][3]+=1
```


# Show all sample length and its type

```python
samples={}
for r ,ds,fs in os.walk(dataset):
    for f in fs:
        path=os.path.join(r,f)
        group=getGroup(f)
        if "vali" not in path and 'txt' not in f and 'jpg' in f:
            if group not in samples:
                samples[group]={'files':[],'result':0,'duration':0,'length':0}
            samples[group]['files'].append(f)
            if 'positive' in path:
                samples[group]['result']=1
            if 'polluted' in path:
                samples[group]['result']=2
```

```python
for k,v in samples.items():
    samples[k]['files']=sorted(samples[k]['files'])
```

```python
for k,v in samples.items(): 
    samples[k]['duration']=(getDate(samples[k]['files'][-1])-getDate(samples[k]['files'][0])).days+1
    samples[k]['length']=len(samples[k]['files'])
```
