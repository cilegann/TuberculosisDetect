```python
import numpy as np
import os

cnn_prob_file='../result/cnn_prob.csv'
tscnn_prob_file='../result/tscnn_prob.csv'
yolo_prob_file='../result/yolo_prob.csv'
tsyolo_prob_file='../result/tsyolo_prob.csv'
```

```python
def read_probs():
    probs={}
    
    with open(cnn_prob_file) as f:
        next(f)
        lines=f.readlines()
    for line in lines:
        data=line.replace("\n","").split(",")
        fn=data[0].split("/")[4].replace(".txt","").replace(".jpg","")
        if fn not in probs:
            probs[fn]={}
            probs[fn]["real"]=float(data[1])
        probs[fn]["cnn_prob"]={}
        for i in range(3):
            probs[fn]["cnn_prob"][i]=float(data[i+2])
            
    with open(tscnn_prob_file) as f:
        next(f)
        lines=f.readlines()
    for line in lines:
        data=line.replace("\n","").split(",")
        fn=data[0].split("/")[4].replace(".txt","").replace(".jpg","")
        if fn not in probs:
            probs[fn]={}
            probs[fn]["real"]=float(data[1])
        probs[fn]["tscnn_prob"]={}
        for i in range(3):
            probs[fn]["tscnn_prob"][i]=float(data[i+2])
    
    with open(yolo_prob_file) as f:
        next(f)
        lines=f.readlines()
    for line in lines:
        data=line.replace("\n","").split(",")
        fn=data[0].split("/")[4].replace(".txt","").replace(".jpg","")
        if fn not in probs:
            probs[fn]={}
            probs[fn]["real"]=float(data[1])
        probs[fn]["yolo_prob"]={}
        for i in range(3):
            probs[fn]["yolo_prob"][i]=float(data[i+2])
            
    with open(tsyolo_prob_file) as f:
        next(f)
        lines=f.readlines()
    for line in lines:
        data=line.replace("\n","").split(",")
        fn=data[0].split("/")[4].replace(".txt","").replace(".jpg","")
        if fn not in probs:
            probs[fn]={}
            probs[fn]["real"]=float(data[1])
        probs[fn]["tsyolo_prob"]={}
        for i in range(3):
            probs[fn]["tsyolo_prob"][i]=float(data[i+2])
    
    return probs
```

```python
probs=read_probs()
```
